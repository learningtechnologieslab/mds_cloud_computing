```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create a graph
G = nx.Graph()

# Example data: games and their tags
games = {
    "Counter-Strike": ["Action", "Multiplayer", "Shooter", "Tactical"],
    "The Witcher 3": ["RPG", "Singleplayer", "Story-Rich", "Fantasy"],
    "Apex Legends": ["Action", "Shooter", "Multiplayer", "Battle Royale"],
    "Portal 2": ["Adventure", "Puzzle", "Story-Rich", "Co-op"],
    "Rocket League": ["Multiplayer", "Sports", "Action", "Competitive"],
    "Stardew Valley": ["Simulation", "Farming", "Singleplayer", "Indie"],
    "Dark Souls 3": ["RPG", "Action", "Fantasy", "Difficult"],
    "Civilization VI": ["Strategy", "Turn-Based", "Multiplayer", "Historical"],
    "Among Us": ["Multiplayer", "Party", "Co-op", "Indie"],
    "Overwatch": ["Action", "Shooter", "Multiplayer", "Hero-Based"]
}

# Add nodes for games and tags
for game, tags in games.items():
    G.add_node(game, type="game")
    for tag in tags:
        G.add_node(tag, type="tag")
        G.add_edge(game, tag)

# Define colors for visualization
node_colors = ["blue" if G.nodes[node]["type"] == "game" else "green" for node in G.nodes]

# Draw the graph
plt.figure(figsize=(12, 9))
pos = nx.spring_layout(G, seed=42)  # Layout for the graph
nx.draw(
    G, 
    pos, 
    with_labels=True, 
    node_color=node_colors, 
    node_size=1500, 
    font_size=10, 
    font_color="white",
    edge_color="gray"
)
plt.title("Steam Games and Tags Graph", fontsize=16)
plt.show()

# Example query: Find games that share the same tags as "Counter-Strike"
query_game = "Counter-Strike"
query_tags = set(games[query_game])

# Find other games that share at least one tag with the query game
related_games = set()
for game, tags in games.items():
    if game != query_game and query_tags.intersection(tags):
        related_games.add(game)

print(f"Games sharing tags with {query_game}: {related_games}")
```
